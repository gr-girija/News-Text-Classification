```python
import re
import csv
import operator
import nltk
from nltk import PorterStemmer
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download("all")
```

```python
paragraph = "Delhi News Live: IYC protests in Delhi demanding arrest of SFI members who vandalised Rahul Gandhi’s Wayanad office"

res = re.sub(r'[^\w\s]', '', paragraph) #removing all special characters

sentence = nltk.sent_tokenize(res)

stemmed=[]
stemmer = PorterStemmer()

for i in range(len(sentence)):
  words = nltk.word_tokenize(sentence[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  stemmed.append(' '.join(words))

lemmatized=[]
lemm=WordNetLemmatizer()

for i in range(len(sentence)):
  words = nltk.word_tokenize(sentence[i])
  words = [lemm.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  lemmatized.append(' '.join(words))

def bag_of_words(sentence):
  bag={}
  for i in sentence:
    words = i.split()
    for word in words:
      if word in bag:
        bag[word]+1
      else:
        bag[word]=1
  return bag


stembag = bag_of_words(stemmed)
lemmbag = bag_of_words(lemmatized) 

print("stemmed bag: ",stembag)
print("lemmatized bag: ",lemmbag)

x = {w for w in list(stembag.keys())}
y = {w for w in list(lemmbag.keys())}


l1=[]
l2=[]

rvec = x.union(y)

for w in rvec:
  if w in x:
    l1.append(1)
  else:
    l1.append(0)

  if w in y:
    l2.append(1)
  else:
    l2.append(0)

c=0
for i in range(len(rvec)):
  c+=l1[i]*l2[i]             

cosine = c/float((sum(l1)*sum(l2))**0.5)

print("similarity between lemmatization and stemming: ",cosine,"or",cosine*100,"%")
```
